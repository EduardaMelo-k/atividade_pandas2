## Passo 0 — Importar pandas e carregar os dados

```python
import pandas as pd
df = pd.read_parquet('emendas_parlamentares (1).parquet')
df.head()
```

## 1) Filtre apenas as emendas do ano de **2022** e mostre as 30 primeiras linhas. Em seguida responda: Quanta emendas foram registradas em 2022?

```python
df_2022 = df[df['Ano da Emenda'] == 2022]
df_2022.head(30)
df_2022.shape[0]
```

Foram registradas 6108 emendas no ano de 2022. 

## 2) Filtre as emendas da região **Norte** que tenham `Valor Empenhado` maior que **500.000** e mostre as 5 primeiras linhas. Em seguida, responda: quantas emendas atendem a esses duas condições ao mesmo tempo?

```python
filtro = (df['Localidade de aplicação do recurso'].str.contains(' - PA| - AM| - RO| - AC| - RR| - AP| - TO')) & (df['Valor Empenhado'] > 500000)
df_norte_500k = df[filtro]

df_norte_500k.head(5)
df_norte_500k.shape[0]
```

2067 emendas atendem as condições de localidade Região Norte e valor empenhado maior que 500.000 ao mesmo tempo.

## 3) Filtre as emendas que são do estado de **SÃO PAULO** ou do estado do **RIO DE JANEIRO** e mostre quantas linhas cada estado possui. Em seguida, responda: Qual dos dois estados possui mais emendas e quantas são?

```python
df_sp = df[df['Localidade de aplicação do recurso'].str.contains(' - SP')]
df_rj = df[df['Localidade de aplicação do recurso'].str.contains(' - RJ')]

len(df_sp)
len(df_rj)

df[df['Localidade de aplicação do recurso'].str.contains(' - SP| - RJ')]['Localidade de aplicação do recurso'].str.extract('(- ..)')[0].value_counts()
```

O Estado que possui mais emendas é São Paulo, contendo 4730 emendas.

## 4) Na coluna `Código da Emenda`, substitua todos os valores `"Sem informação"` por `"Não disponível"`. Mostre a contagem de valores únicos da coluna antes e depois da substituição.

```python
print(df['Código da Emenda'].value_counts())


df['Código da Emenda'] = df['Código da Emenda'].replace('Sem informação', 'Não disponível')

print(df['Código da Emenda'].value_counts())
```

A contagem de valores únicos da coluna permanece sendo 17.808.

## 5) Na coluna `Nome Função`, transforme todos os valores para que fiquem em letras maiúsculas. Confirme a alteração mostrando os valores únicos da coluna.

```python
df['Nome Função'] = df['Nome Função'].str.upper()
df['Nome Função'].unique()
```

Utilizando o método str.upper() é possível transformar todos os valores disponíveis na coluna Nome Função para que fiquem em letra maiúscula. 

## 6) Verifique quantas linhas duplicadas existem no dataset. Em seguida, remova as duplicatas e mostre quantas linhas sobraram.

```python
print(df.duplicated().sum())
df = df.drop_duplicates()
print(df.shape[0])
```

Não foram encontradas linha duplicadas no dataset. O número total de linhas é de 88596.

## 7) Remova duplicatas considerando apenas as colunas `Código da Emenda` e `Município`. Quantas linhas sobraram? O resultado é diferente de quando removemos duplicatas do DataFrame inteiro? 

```python
print(df.duplicated(subset=['Código da Emenda', 'Município']).sum())
df_subset = df.drop_duplicates(subset=['Código da Emenda', 'Município'])
print(df_subset.shape[0])
```

Existem 12852 linhas duplicadas considerando apenas o par Código da Emenda e Município. Sobraram 75744 linhas após a remoção.

## 8) Identifique quais colunas possuem valores nulos e quantos nulos cada uma tem. Em seguida, responda: qual é a coluna que possui maior número de valores nulos e quantos são?

```python
nulos_por_coluna = df.isnull().sum()
print(nulos_por_coluna)

coluna_max_nulos = nulos_por_coluna.idxmax()
total_max_nulos = nulos_por_coluna.max()
print(coluna_max_nulos, total_max_nulos)
```

As colunas Código do Autor da Emenda e Nome do Autor da Emenda são as que possuem maior volume de dados faltantes no arquivo, com 15.961 linhas vazias. Todas as colunas que aparecem com 0 estão totalmente preenchidas.

## 9) Preencha os valores nulos da coluna `codigo do autor` com **-1** e os valores nulos da coluna `nome do autor da emenda` com "desconhecido" e confirme que não restaram nulos nessa coluna.

```python
df['Código do Autor da Emenda'] = df['Código do Autor da Emenda'].fillna(-1)
df['Nome do Autor da Emenda'] = df['Nome do Autor da Emenda'].fillna('desconhecido')

print(df[['Código do Autor da Emenda', 'Nome do Autor da Emenda']].isnull().sum())
```

os 15.961 espaços que antes estavam vazios agora contêm -1 (no código) ou "desconhecido" (no nome). O dataset está preenchido nessas colunas.

## 10) Verifique o tipo da coluna `Ano da Emenda`. Converta-a para **string** (`str`). Mostre o tipo antes e depois da conversão.

```python
print(df['Ano da Emenda'].dtype)
df['Ano da Emenda'] = df['Ano da Emenda'].astype(str)
print(df['Ano da Emenda'].dtype)
```

Antes: Os números eram tratados como números inteiros.
Depois: Após convertido para String, o tipo aparece como object. Isso significa que o ano é tratado como um texto.

## 11) Converta a coluna `Valor Empenhado` para o tipo **float** (caso ainda não seja). Mostre o tipo após a conversão.

```python
df['Valor Empenhado'] = df['Valor Empenhado'].astype(float)
print(df['Valor Empenhado'].dtype)
```

Após a conversão, o tipo da coluna Valor Empenhado é float.

## 12) Calcule o **total** de `Valor Pago` por **Região**. Quais são as duas regiões que mais receberam recursos?

```python
total_por_regiao = df.groupby('Região')['Valor Pago'].sum().sort_values(ascending=False)
print(total_por_regiao)

print("\nAs duas regiões que mais receberam recursos:")
print(total_por_regiao.head(2))
```

As duas regiões que mais receberam recursos foram Nordeste e Sudeste.
